```python
import pandas as pd
import json
```

```python
# load json

with open('top_movies.json') as json_file:
    data = json.load(json_file)
 
data
```

```python
actor = []
movie = []
for i in data:
    for j in i["cast_list"]:
        actor.append(j)
        movie.append(i['title'])
        
    
```

```python
len(actor)
```

```python
actor_movie = {}
for i in range(len(actor)):
    if actor[i] in actor_movie.keys():
        actor_movie[actor[i]].append(movie[i])
    else:
        actor_movie[actor[i]] = [movie[i]]
```

```python
actor_movie_num ={}
for i in actor_movie:
    actor_movie_num[i] = len(actor_movie[i]) 
```
